```python
import pandas as pd
```

```python
tb_one = pd.read_csv("test2_1.csv")
tb_two = pd.read_csv("test2_3.csv")
```

```python
tb_one_new = tb_one.groupby("Doc.Num")["Invoice Quantity",  "Cost Price",'Taxable Amount', 'Total Tax', 'Total Amount','MRP Price'].sum().reset_index()
tb_one_new["Itm_no"] = tb_one.groupby("Doc.Num")['Itm.no'].unique().values
tb_one_new["Currency"] = tb_one.groupby("Doc.Num")['Currency'].unique().values
tb_one_new["Bill_date"] = tb_one.groupby("Doc.Num")['Bill Date'].unique().values
tb_one_new["Payer City"] = tb_one.groupby("Doc.Num")['Payer City'].unique().values
```

```python
tb_two_new = tb_two.groupby("Doc.Num")["Invoice Quantity", "Cost Price",'Taxable Amount', 'Total Tax', 'Total Amount','MRP Price'].sum().reset_index()
tb_two_new["Itm_no"] = tb_two.groupby("Doc.Num")['Itm.no'].unique().values
tb_two_new["Currency"] = tb_two.groupby("Doc.Num")['Currency'].unique().values
tb_two_new["Bill_date"] = tb_two.groupby("Doc.Num")['Bill Date'].unique().values
tb_two_new["Payer City"] = tb_two.groupby("Doc.Num")['Payer City'].unique().values
```

```python
tb_output = tb_one_new.merge(tb_two_new, on=['Doc.Num'])
```
